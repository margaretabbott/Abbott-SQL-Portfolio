## Problem 1

List the name and weight of finished products that weigh more than 2 pounds. For the weight, output a single column named **ProductWeight** that includes both the weight and unit(e.g., **2.68 LB** or **1050.00 G**).

**Hint**:
- Use the **Product** table.
- A finished prodcut's *FinishedGoodsFlag* is set to 1.
- Check the unique values in the *WeightUnitMeasureCode* column to understand the unit of measure for each produt's weight.
- 1 lb = 453.59 grams

```sql
-- Your SQL query Here
SELECT [Name], CAST([Weight] AS nvarchar) + ' ' + WeightUnitMeasureCode AS ProductWeight
FROM Production.Product
WHERE FinishedGoodsFlag=1 AND ((WeightUnitMeasureCode='LB' AND [Weight]>2) OR (WeightUnitMeasureCode='G' AND [Weight]>2*453.59));
```

## Problem 2

The Marketing Department is preparing an email promotion that will be sent to all individuals who opted in. Provide a list of first names, last names, and email addresses for those eligible.

**Hint**:
- Use the **EmailAddress** and **Person** tables.
- A person is considered opted in when his/her *EmailPromotion* value is 1 or 2.

```sql
-- Your SQL query here
SELECT FirstName, LastName, ea.EmailAddress
FROM Person.Person AS per
INNER JOIN Person.EmailAddress AS ea 
ON per.BusinessEntityID=ea.BusinessEntityID
WHERE EmailPromotion=1 OR EmailPromotion=2
```

## Problem 3


List the first names, last names, email addresses, andÂ password modified dates (renamed as **PasswordModifiedDate**) of current employees who have not changed their passwords since January 1, 2008. Sort the result by password modified dates in ascending order.

**Hint**:
- Use the **Person**, **EmailAddress**, **Password**, and **Employee** tables.
- The *ModifiedDate* column in the **Password** records the last date and time an employee modified his/her password.
- A current employee's *CurrentFlag* is set to 1.

```sql
-- Your SQL query here
SELECT FirstName, LastName, ea.EmailAddress, pas.ModifiedDate AS PasswordModifiedDate
FROM Person.Person AS per 
INNER JOIN Person.EmailAddress AS ea 
ON per.BusinessEntityID=ea.BusinessEntityID
INNER JOIN Person.[Password] AS pas 
ON per.BusinessEntityID=pas.BusinessEntityID
INNER JOIN HumanResources.Employee AS hr 
ON per.BusinessEntityID=hr.BusinessEntityID
WHERE DATENAME(YEAR, pas.ModifiedDate)<2008 AND CurrentFlag=1
ORDER BY pas.ModifiedDate;
```

## Problem 4

The Sales Manager wants to know which products were sold with more than 50% discount.Â Create a report showing each unique product name, list price, sold price, and the discount (named as **Discount**). Sort the result by Discount in descending order.

**Hint**:
+ The *ListPrice* column in the **Product** table shows a product's list price.
+ The *UnitPrice* column in the **SalesOrderDetail** table shows the price at which a product was sold.
+ Discount = (ListPrice - UnitPrice) / ListPrice

```sql
-- Your SQL query here
SELECT DISTINCT [Name], ListPrice, sod.UnitPrice, ((ListPrice-sod.UnitPrice)/ListPrice)*100 AS Discount
FROM Production.Product AS pp
INNER JOIN Sales.SalesOrderDetail AS sod
ON pp.ProductID=sod.ProductID
WHERE (((ListPrice-sod.UnitPrice)/ListPrice)*100)>50
ORDER BY Discount DESC;
```

## Problem 5

The HR Department wants a list of all employees currently working in the **Marketing** department who were hired before 2008 or after 2010. Provide a list of eligible employees, including their BusinessEntityID, job titles, department name, and hire dates. Sort the results by hire dates in ascending order.

**Hint**:
+ An employee is currently working in a department only when his/her *EndDate* is NULL.
+ The *Name* column in the **Department** table shows a department's name.


```sql
-- Your SQL query here
SELECT hre.BusinessEntityID, JobTitle, hrd.[Name], HireDate
FROM HumanResources.Employee AS hre
INNER JOIN HumanResources.EmployeeDepartmentHistory AS edh
ON hre.BusinessEntityID=edh.BusinessEntityID
INNER JOIN HumanResources.Department AS hrd
ON edh.DepartmentID=hrd.DepartmentID
WHERE hrd.Name='Marketing' AND edh.EndDate IS NULL AND (DATENAME(YEAR, hre.HireDate)>2010 OR DATENAME(YEAR, hre.HireDate)<2008)
ORDER BY hre.HireDate;
```

The Production Manager wants a list of finished produts that are still for sale but have never been sold. Provide a list of the eligible product names in alphabetical order.

**Hint**:
+ Finished prodcuts that are still for sale can be identified by using `FinishedGoodsFlag = 1` and `SellEndDate is NULL`.
+ Products that have never sold don't exist in the *SalesOrderDetail* table.


```sql
-- Your SQL query here
SELECT [Name]
FROM Production.Product AS pp
LEFT JOIN Sales.SalesOrderDetail AS sod 
ON pp.ProductID=sod.ProductID
WHERE FinishedGoodsFlag=1 AND SellEndDate IS NULL AND sod.ProductID IS NULL 
ORDER BY [Name];
```
